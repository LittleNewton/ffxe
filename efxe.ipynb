# Embedded FXE Example with Blinky

testing out the embedded FXE engine using nRF52832 blinky firmware example

```python
from efxe import *
```

```python
efxe = EmFXEngine(
    pd="mmaps/nrf52832.yml",
    fw_path="examples/blinky.bin",
    log_stdout=True,
    log_insn=True,
)
```

```python
efxe.run()
```

```python
efxe.print_regs()
```

```python
efxe.uc.emu_start(efxe.context.pc, 0x80000, count=5)
```

```python
for fn_addr, fp, ret_addr in efxe.context.callstack:
    print("fn: {:<10} fp: {:<10} ret: {:<10}".format(
        hex(fn_addr), hex(fp), hex(ret_addr) if ret_addr else str(None)))
```

```python
print('\n'.join(efxe.fw.disasm_txt))
```
